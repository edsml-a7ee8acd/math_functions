```python
import numpy as np
import pylab
from sympy import sin, cos, exp, Function, Symbol, diff, integrate, solve
from math import pi
import matplotlib.mlab as mlab
```

**SCALAR FIELDS**


```python
x = np.arange(-5, 5, 0.025)
y = np.arange(-5, 5, 0.025)
x, y = np.meshgrid(x, y)
```

```python
#The above creates a mesh of 2D Cartesian coordinates
#From which, the value of the scalar field can be calculated
z = 26 - 2*(x**2) - y**3 + x*y
```

```python
#Plots the scalar field on a contour plot 
fig = pylab.figure()
contour_plot = pylab.contour(x, y, z)
pylab.clabel(contour_plot,inline=1)
pylab.xlabel('x')
pylab.ylabel('y')
print("Plotting contours of the scalar field f(x,y) = 16 - 2*(x**2) - y**2 + x*y")
```

**VECTOR FIELDS**

```python
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)

#the meshgrid function creates a 2D grid from the
#given x, y ranges
```

```python
z1 = -(x**2)
z2 = -(y**2)

#z1 and z2 hold the i and j components of the vector
#field respectively 
```

```python
fig = pylab.figure()
plt = pylab.quiver(x, y, z1, z2, angles='xy', scale=1000, color='r')
pylab.quiverkey(plt, -5, 5.5, 50, '50 units', coordinates='data', color='r')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.show()

print("Plotting the vector field f(x,y) = [-x**2, -y**2]")
```

```python
#The quiver command from pylab plots arrows in the x,y plane based
#on the magnitude and unit direction of z1, z2 
```

**GRADIENTS**

```python
x = Symbol('x')
y = Symbol('y')
f = 16 - 2*(x**2) - y**2 + x*y
```

```python
grad_f = [diff(f,x), diff(f,y)]
print(solve([grad_f[0], grad_f[1]], [x,y]))
```

**DIRECTIONAL DERIVATIVES**

Given by:


```python
x = Symbol('x')
y = Symbol('y')
h = 3 * x * (y**2)
```

```python
grad_h = [diff(h,x), diff(h,y)]
```

```python
grad_h_at_point = [grad_h[0].evalf(subs={x:1, y:2}), grad_h[1].evalf(subs={x:1, y:2})]
```

```python
print(grad_h_at_point)
```

```python
a = np.array([3, 4])

#sqrt (a[0]**2 + a[1]**2)
a_magnitude = np.linalg.norm(a, ord=2)

#Each vector component divided by its magnitude
unit_a = a/a_magnitude
```

```python
slope = np.dot(grad_h_at_point, unit_a)
```

```python
print("The slope of h in the direction ", unit_a, " at (1,2) is: ", slope)
```

**DIVERGENCE**

Given by:


```python
x = Symbol('x')
y = Symbol('y')
```

```python
v = [-(x**2), -(y**2)]
```

```python
divergence = diff(v[0], x) + diff(v[1], y)
#d/dx applied to the first component of v, v[0]
#d/dy applied to the second component of v, v[1]
```

```python
print(divergence)
```

**CURL**

Given by:


```python
x = Symbol('x')
y = Symbol('y')
v = [cos(pi*y), -cos(pi*x)]
```

```python
#The first two components are zero because v has 
#a zero k-component
curl = [0, 0, diff(v[1], x) - diff(v[0], y)]
#The function differentiates the 2nd component in x
#and the 1st component in y 
```

```python
print(curl)
```

```python
print([0, 0, curl[2].evalf(subs={x:0, y:-0.5})])
```

**LAPLACIAN**

Given by:


```python
x = Symbol('x')
y = Symbol('y')
f = x*y + 3*exp(x*y) #defines the scalar field x,y
```

```python
laplacian = diff(f, x, 2) + diff(f, y, 2)
#the '2' in the diff equation specifies the order 
#of the equation. This is 2nd order due to the 
#equation for the laplacian equation given above.
print(laplacian)
```
