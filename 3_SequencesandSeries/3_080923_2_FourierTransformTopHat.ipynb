```python
#Define the top hat function for a = 1
def top_hat(t):
    if -0.5< t <0.5:
        z = 1
    else:
        z = 0
    return z

t = np.linspace(-1, 1, 200)
f_t = []
for i in t:
    y = top_hat(i)
    f_t.append(y)
    
omega = np.fft.fftfreq(len(t), d=t[1] - t[0])
F_omega = np.fft.fft(f_t)

plt.subplot(1,2,1)
plt.plot(t, f_t)
plt.title("Top hat function")

plt.subplot(1,2,2)
plt.title("Add title")
plt.plot(omega, np.real(F_omega), '-')
plt.tight_layout()
plt.show()
```

```python
t = np.linspace(-1, 1, 1000)
omega = np.linspace(-50, 50, 1000)

f_t = []
for i in t:
    y = top_hat(i)
    f_t.append(y)
F_omega = []
for i in omega:
    z = np.sinc(i/(2*np.pi))
    F_omega.append(z)
    
plt.subplot(1,2,1)
plt.plot(t, f_t)
plt.title("Top hat function")

plt.subplot(1,2,2)
plt.title("Add title")
plt.plot(omega, F_omega)

plt.tight_layout()
plt.show()
```
