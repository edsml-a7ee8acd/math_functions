```python
#the first package is a package from which the Lagrange polynomial can be generated
import scipy.interpolate as si
import numpy as np
from matplotlib import pyplot as plt
```

**Define a function plot_raw_data which inputs the established axes, x and y datasets and plots them**

```python
def plot_raw_data(x, y, ax):
    ax.plot(x, y, 'ko', label='raw data')
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('y', fontsize=12)
    ax.grid(True)
```

```python
#a set of raw data for a matrix system x and y
xi = np.array([0.5, 2.0, 4.0, 5.0, 7.0, 9.0])
yi = np.array([0.5, 0.4, 0.3, 0.1, 0.9, 0.8])
```

```python
#the lagrange polynomial for the datasets xi and yi
lp = si.lagrange(xi, yi)
```

```python
#Evaluate the function at a high resolution
x = np.linspace(0.4, 9.1, 100)
```

```python
# set up the figure
fig = plt.figure(figsize=(7, 5))
ax1 = fig.add_subplot(111)
ax1.margins(0.1)

# Actually plot (x,y)=(x,lp(x)) on the axes with the label ax1
ax1.plot(x, lp(x), 'b', label='Lagrange interpolating polynomial')

#Apply the raw data algorithm to the axes with the xi and yi axes
plot_raw_data(xi, yi, ax1)
ax1.set_title('Lagrange interpolating polynomial (SciPy)', fontsize=16)
ax1.legend(loc='best', fontsize=14)
plt.show()
```
