```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plot_raw_data(x, y, ax):
    ax.plot(x, y, 'ko', label='raw data')
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('y', fontsize=12)
    ax.grid(True)
```

```python
x = np.array([0.5, 2.0, 4.0, 5.0, 7.0, 9.0])
y = np.array([0.5, 0.4, 0.3, 0.1, 0.9, 0.8])
deg = 2 #1st degree = linear, 2nd degree = quadratic etc.
```

```python
poly_coeffs = np.polyfit(x, y, deg, rcond=None, full=False, w=None, cov=False)
```

```python
p1 = np.poly1d(poly_coeffs) #turns the coefficients into a function which we can evaluate
```

```python
print(p1)
```

```python
fig = plt.figure(figsize=(7, 5))
ax1 = fig.add_subplot(111)
ax1.margins(0.1)

xx = np.linspace(0., 9.5, 100) #set up a range of x values to plot on axes
ax1.plot(xx, p1(xx), 'b', label=r'$y = {0:.4f}x + {1:.4f}$'.format(poly_coeffs[0], poly_coeffs[1]))
plot_raw_data(x, y, ax1)

ax1.legend(loc='best', fontsize=12)

# Add a figure title
ax1.set_title('Raw data and the corresponding linear best fit line', fontsize=12)

plt.show()
```
