```python
import matplotlib.pyplot as plt
import numpy as np
```

The formula for the Forward Euler Scheme is 



This formula is emulated in the u = u + dt x f(u,t) component of the function below

```python
def euler(f, u0, t0, t_max, dt):
    u = u0
    t = t0
    #store all solution values and associated time levels for later plotting
    u_all = [u0]
    t_all = [t0]
    while t<t_max:
        u = u + dt*f(u,t)
        u_all.append(u)
        t = t + dt
        t_all.append(t)
    return u_all, t_all
```

```python
def f1(u, t):
    val = u
    return val

def f2(u, t):
    val = np.cos(t)
    return val
```

**Demonstration of function for u'(t) = u(t) and u'(t) = cos(t)**

```python
u1, t1 = euler(f1, 1.0, 0.0, 10.0, 0.1) # time starts at 0 and is incremented by 0.5 after every iteration
#u1,t1 applies the function u'(t) = u(t) to the array formed by the other arguments of this function call

u2, t2 = euler(f2, 1.0, 0.0, 10.0, 0.1) #u2,t2 applies the u'(t) = cos(t) 
```

```python
#set up figure
fig, axes = plt.subplots(1,2, figsize=(10, 4))
ax1 = axes[0]
ax2 = axes[1]

ax1.plot(t1, u1, color="blue",
         label="$u'(t)=u(t)$,\n$u(0)=1$")
ax1.set_xlabel('$t$', fontsize=14)
ax1.set_ylabel('$u(t)$', fontsize=14)
ax1.legend(loc="best", fontsize=12)

ax2.plot(t2, u2, color="blue",
         label="$u'(t)=\cos(t)$,\n$u(0)=0$")
ax2.set_xlabel('$t$', fontsize=14)
ax2.set_ylabel('$u(t)$', fontsize=14)
ax2.legend(loc="lower left", fontsize=12)

plt.subplots_adjust(wspace=0.3)
plt.suptitle("Forward Euler scheme solutions", fontsize=14)
plt.show()
```
