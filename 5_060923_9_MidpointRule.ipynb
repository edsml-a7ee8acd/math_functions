```python
from matplotlib.patches import Rectangle
from matplotlib import pyplot as plt
#this function plots rectangles over a function which 
#helps us to visualise the effect of the midpoint rule
import numpy as np
```

The midpoint rule for a given iteration of x takes the form:



The midpoint estimate of the entire function I then requires summing up all iterations:



This is essentially calculating the area of a given rectangle whose width is (xi+1 - xi) and height is f((xi+1 + xi) / 2)

```python
def f(x):
    #write out the function we wish to integrate. In 
    #this case we will write np.sin(x)
    return np.sin(x)

#equally spaced values to iterate through
x = np.linspace(0, np.pi, 100)
y = f(x)
```

```python
fig = plt.figure(figsize=(10, 4))
ax1 = plt.subplot(111)
ax1.plot(x, y, 'b', lw=2)

ax1.margins(0.1)

# Label axis.
ax1.set_xlabel('x', fontsize=14)
ax1.set_ylabel('$f(x)=\sin(x)$', fontsize=14)

# Overlay a grid.
ax1.grid(True)

number_intervals = 5
xi = np.linspace(0, np.pi, number_intervals+1)
I_M = 0.0
for i in range(number_intervals):
    ax1.add_patch(Rectangle((xi[i], 0.0), (xi[i+1] - xi[i]),
                            f((xi[i+1]+xi[i])/2), fill=False, ls='--', color='k', lw=2))
    I_M += f((xi[i+1]+xi[i])/2)*(xi[i+1] - xi[i])

ax1.set_title('The sum of the areas of the rectangles is $I_M =$ {:.12f}.'.format(I_M),
              fontsize=14)

    
plt.show()
```
