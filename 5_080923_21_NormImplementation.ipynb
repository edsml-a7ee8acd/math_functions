**Code to compute the two matrix norms defined mathematically above**

```python
def frob(A): #Frobenius norm 
    m, n = A.shape()
    squsum = 0
    for i in range(m):
        for j in range(n): # a double sigma can be written as two for-loops
            squsum += A[i,j]**2
    return np.sqrt(squsum)

def mars(A):
    m, n = A.shape()
    maxarsum = 0
    for i in range(m):
        arsum = np.sum(np.abs(A[i]))
        maxarsum = arsum if arsum > maxarsum else maxarsum 
    return maxarsum 

A = np.array([[10, 2, 1], [6, 5, 4], [1, 4, 7]])
```

**Matrix Conditioning**
The conditioning of a matrix is measured with the condition:


If cond(A) is close to one, then A is termed 'well-conditioned', as it approaches infinity, it becomes more 'ill-conditioned' 

```python
A = np.array([[10, 2, 1], [6,5,4], [1,4,7]])

#Definitions of conditioning 
print("A =", A)
print("SciPy cond(A) =", np.linalg.cond(A))
print("Default condition number uses matrix two-norm =", sl.norm(A,2)*sl.norm(sl.inv(A),2))
print("SciPy Frobenius cond(A) = ", np.linalg.cond(A, 'fro'))
```
