```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def central_diff(f, x, dx):
    numerator = f(x+dx) - f(x-dx)
    derivative = numerator / (2.0*dx)
    return derivative

def forward_diff(f, x, dx):
    numerator = (f(x+dx) - f(x))
    derivative = numerator / dx
    return derivative
```

```python
#Apply the two functions to a sine wave to compare their 
#accuracy with a known value 

fd_errors = []
cd_errors = [] #arrays to store errors of each differentiation method
dx_all = []
dx = 1.0 #an initial dx value 
exact = np.sin(0.8)

for i in range(10):
    fd = forward_diff(np.sin, 0.8, dx)
    cd = central_diff(np.sin, 0.8, dx)
    fd_errors.append(abs(fd-exact))
    cd_errors.append(abs(cd-exact))
    dx = dx/2 # we need to halve the derivative so that 
    #the difference between x and dx is reduced 
    #so our estimate gets better
    print('%10g (error=%10.2g) %10g (error=%10.2g)' % (fd, abs(fd-exact), cd, abs(cd-exact)))
```
